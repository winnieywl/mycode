```python
import pandas as pd

excel_file = "movies.xls"
```

```python
movies = pd.read_excel(excel_file) # automatically only scans the first sheet
```

```python
movies.head()
```

```python
movies_sheet1 = pd.read_excel(excel_file, sheet_name=0, index_col=0) # first sheet from that excel file
movies_sheet1.tail()
```

```python
movies_sheet1.head(5).to_excel("5movies.xlsx") # to_excel method belongs to the dataframe
movies_sheet1.head(5).to_json("5movies.json")
movies_sheet1.head(5).to_csv("5movies.csv")
```
