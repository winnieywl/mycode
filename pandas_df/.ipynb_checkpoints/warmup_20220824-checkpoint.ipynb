```python
import pandas as pd
```

```python
cars1 = pd.read_csv("cars1.csv", index_col = 0)
cars2 = pd.read_csv("cars2.csv", index_col = 0)
cars = pd.concat([cars1, cars2])
```

```python
cars
```

# Top 10 Most Expensive Cars

```python
# Top 10 Most Expensive Cars
cars[["VIN", "Model", "Price", "timestamp"]].sort_values("Price", ascending = False).head(10)
# [["Stock #", "VIN", "Model", "Price", "timestamp"]]
```

Find minimum values: Stack Overflow solutions:        
https://stackoverflow.com/questions/34014137/multiple-minimum-values-in-pandas-dataframe

```python
# least amount of miles
cars[cars['Odometer'] == cars['Odometer'].min()]
```

```python
# sum of prices 
cars["Price"].sum()
print("If all cars were damaged, the total loss is", round(cars["Price"].sum()), "dollars.")
```

```python
cars_withphotos = cars[cars['Photo Count'] != 0]
cars_withphotos
```
